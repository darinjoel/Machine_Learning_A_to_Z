## Data Preprocessing


```python
# Data Preprocessing Template

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values    # Only 1 ind col, so could also say col 0
y = dataset.iloc[:, 1].values # Grabbing second col as dependent variable

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0) ## Using 1/3 per tutorial
```

## Create and train model

```python
# Fitting Simple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() # Create regressor object from LinearRegression class
regressor.fit(X_train, y_train) # Fit regressor object to the data using 'fit' method.
```

## Predict test set using model

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test) # call predict function from regressor class to return predicted
```

## Post Processing

```python
# Visualizing the Training set results
plt.scatter(X_train, y_train, color = 'red')

'''
Plotting the regression line. This is in effect 'the model'. We are plotting what the regressor
object that we have trained on our training data predicts for the training data.

'''
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# Visualizing the Test set results
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

```

## Lets look at y_pred further

```python
# Lets examine y_pred
"""
We could instead plot the points from y pred, and they should fall exactly on the regressor line that we trained from our 
training data
"""
plt.scatter(X_test, y_pred, color = 'black', marker='h', s=100)
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (y_pred)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

"""
Lets plot error
"""

error = ((y_pred - y_test)/y_test )*100

plt.scatter(X_test, error, color = 'green')
plt.title('Error vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
